# Texas

## * The datasets are obtained from GOSAT; dates from 2017 to 2020(span of 3 years) *

```python
import os
os.listdir()   # Show me what is in the directory
```

```python
data_blue = pandas.read_excel("Excel Data/Texas_CH4.xls")
```

```python
data_blue.head()
```

```python
print(data_blue['Dates'])
```

```python
plt.figure(figsize=(15,7))   # Set the size of your plot. It will determine the relative size of all the labels.
plt.xticks(rotation=90)
plt.plot(data_blue['Dates'],data_blue['CH4_apr'],label="Blue Led")
plt.plot(data_blue['Dates'],data_blue['CH4_low'],label="Red")
plt.plot(data_blue['Dates'],data_blue['CH4_upp'],label="Green")# Plot a curve.
plt.xlabel("Dates")
plt.ylabel("CH4_apr, CH4_low, CH4_upp (IN PPMV)")
plt.title("Texas_CH4")
plt.show()
```

### ● The blue line represent the priori of the column averaged dry air mole fraction of CH4 over the period of 3 years

### ● Red line shows the CH4 levels in lower troposphere

### ● Green line shows the CH4 levels in upper troposphere
